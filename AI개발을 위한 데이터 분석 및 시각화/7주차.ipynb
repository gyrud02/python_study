```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

for row in data:
    if '주례제2동' in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
# import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

for row in data:
    if '주례제2동' in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
plt.barh(range(101), result) # barh() : 그래프 수평
plt.show()
```

### 2. 항아리 모양 그래프 그리기
##### - 데이터 수집하기 https://www.mois.go.kr

```python
import csv
f = open('7_gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '주례제2동' in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
        
f.reverse() # reverse() : 반전함수
print([m])
print([f])
```

```python
import csv
f = open('7_gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '주례제2동' in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for i in row[106:]:
            f.append(int(i))
        
print([m])
print([f])
```

```python
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import csv
f = open('7_gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '주례제2동' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
        
print([m])
print([f])

plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

### 3. pie() 함수
#### * 옵션추가하기
##### - matplotlib 라이브러리에서 사용할 수 있는 다양한 색의 이름
https://matplotlib.org/gallery/color/named_colors.html

```python
import matplotlib.pyplot as plt
plt.pie([10,20])
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]
plt.axis('equal')
plt.pie(size)
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
size = [2441, 2312, 1031, 1233]
label = ['A','B','O','AB']
plt.axis('equal')
plt.pie(size, labels = label)
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
size = [2441, 2312, 1031, 1233]
label = ['A','B','O','AB']
plt.axis('equal')
plt.pie(size, labels=label, autopct = '%.1f%%')
plt.legend()
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
size = [2441, 2312, 1031, 1233]
label = ['A','B','O','AB']
color = ['darkmagenta', 'deeppink', 'yellow', 'skyblue']
plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%', explode = (0,0,0.1,0), colors = color)
# autopct = 퍼센트 차지율 / explode = 그래프 돌출 0은 돌출X
plt.legend()
plt.show()
```

### 5. 버블 차트
##### - matplotlib 라이브러리에서 사용할 수 있는 다양한 컬러맵의 종류

```python
plt.style.use('ggplot')
plt.scatter([1,2,3,4], [10,30,20,40], s = [100,200,250,300])
# s = 점의 크기
plt.show()
```

```python
plt.style.use('ggplot')
plt.scatter([1,2,3,4], [10,30,20,40], s = [100,200,250,300], c = ['red', 'coral', 'yellow', 'pink'])
# s = 점의 크기 / c = 점의 색상
plt.show()
```

```python
plt.style.use('ggplot')
plt.scatter([1,2,3,4], [10,30,20,40], s = [100,200,250,300], c = range(4))
# s = 점의 크기 / c = 점의 색상
plt.colorbar() # 오른쪽 컬러 막대
plt.show()
```

```python
plt.style.use('ggplot')
plt.scatter([1,2,3,4], [10,30,20,40], s = [100,200,250,300], c = range(4), cmap = 'prism')
# s = 점의 크기 / c = 점의 색상 / cmap(컬러맵) = jet(무지개색컬러바)
plt.colorbar() # 오른쪽 컬러 막대
plt.show()
```

```python
import random
x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
plt.style.use('ggplot')
plt.scatter(x,y,s = size)
plt.show()
```

```python
plt.scatter(x, y, s = size, c = size, cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x, y, s = size, c = size, cmap = 'jet', alpha = 0.7)
# alpha = 점의 투명도
plt.colorbar()
plt.show()
```
