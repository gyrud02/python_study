### 1. matplotlib 라이브러리
##### - matplotlib 라이브러리는 파이썬에서 2D형태 그래프, 이미지 등을 그릴 때 사용한다. 실제 과학 컴퓨팅 연구 분야. 인공지능 연구에 많이 활용된다.
##### - 파이썬으로 데이터를 시각화하는데는 matplotlib라이브러리를 가장 많이 사용한다.
##### - matplotlib 라이브러리 안에는 다양한 모듈들이 있는데, 그 중 pyplot이라는 모듈을 주로 사용한다.
##### - matplotlib.pyplot은 이름이 너무 길고 복잡하여 plt라는 alias를 사용한다.
##### - plot()에 두 개의 리스트가 ','로 구분되어있다.
##### - 두 개의 리스트를 입력했을 때, 첫번째 리스트가 x축, 두번째 리스트가 y축이다.

```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4], [12,43,25,15])
plt.show()
```

#### plot 함수에 기본 그래프를 그리는 단계
##### 1. import matplotlib.pyplot as plt : 라이브러리 불러오기
##### 2. plt.plot(x축, y축) : plot()에 데이터 입력하기
##### 3. plt.show() : 그래프 보여주기

### 2. 그래프 옵션 추가하기
##### - title()로 그래프에 제목 넣기

```python
plt.title('plotting')
plt.plot([10,20,30,40])
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
plt.title('한글')
plt.plot([10,20,30,40])
plt.show()
```

##### - 그래프에 범례 넣기

```python
plt.rc('font', family='Malgun Gothic')
plt.title('범례넣기')
plt.plot([10,20,30,40], label='상승')
plt.plot([40,30,20,10], label='하강')
plt.legend()
plt.show()
```

##### - 그래프 색상 바꾸기

```python
plt.rc('font', family='Malgun Gothic')
plt.title('색상넣기')
plt.plot([10,20,30,40], color='skyblue', label='상승')
plt.plot([40,30,20,10], color='pink', label='하강')
plt.legend()
plt.show()
```

##### - 그래프 선 모양 바꾸기

```python
plt.rc('font', family='Malgun Gothic')
plt.title('색상넣기')
plt.plot([10,20,30,40], color='skyblue', linestyle='--', label='상승')
plt.plot([40,30,20,10], 'pink', ls=':', label='하강')
plt.legend()
plt.show()
```

##### - 마커 모양 바꾸기

```python
plt.rc('font', family='Malgun Gothic')
plt.title('색상넣기')
plt.plot([10,20,30,40], 'r.', label='상승')
plt.plot([40,30,20,10], 'g^', label='하강')
plt.legend()
plt.show()
```

### 5. 날짜 데이터 추출하기
##### - 문자열 함수 중 split()는 사용자가 설정하는 특정 문자를 기준으로 문자열을 분리하는 함수다.
##### - split()는 사용자가 설정하는 특정 문자가 없다면 기본적으로 공백 문자를 기준으로 문자열을 분리한다.
##### - 사용자가 분리할 문자를 직접 지정할 경우. 공백이 아닌 해당 문자를 기준으로 문자열을 분리한다.
##### - ex) '1907-10-01' 같은 형태의 날짜 데이터를 '-' 기준으로 분리할 수 있다.

##### - split()를 사용하여 1년 중 여름의 정점인 8월의 최고 기온 데이터를 추출하여 그래프로 시각화해보자

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))
            
# print(result)
# f.close

plt.plot(result, 'hotpink')
plt.show()
```

##### - 매년 돌아오는 특정 날짜를 기준으로 그래프 그리기
##### - 매년 2월 14일의 최고 기온 데이터를 추출하여 그래프로 나타내기

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            result.append(float(row[-1]))

# print(result)
# f.close()

plt.plot(result, 'coral')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = [] # 최고 기온값
low = [] # 최저 기온값

for row in data:
    if row[-1] != '' and row[-2] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
            high.append(float(row[-1]))
            low.append(float(row[-2]))

# print(result)
# f.close()

plt.plot(high, 'coral')
plt.plot(low, 'yellow')

plt.rc('font', family = 'Malgun Gothic')
plt.title('기온 변화 그래프')
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```
