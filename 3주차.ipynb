### 판다스 라이브러리 : 파이썬에서 가장 널리 사용되는 데이터 분석 라이브러리로 데이터 프레임(DataFrame)이라는 자료구조다

##### * 데이터 프레임(DataFrame) : 엑셀에서 볼 수 있는 시트(Sheet)와 동일한 개념이다. 시리즈(Series)는 시트의 열 1개를 의미한다.

```python
import pandas as pd # 판다스 라이브러리 import
names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel', 'Henry'] # 변수 names 설정
births = [1990, 1988, 1987, 1978, 2000, 2020] # 변수 births 설정
custom = [1, 5, 6, 7, 8, 10] # 변수 custom 설정
```

```python
DataSet = list(zip(names, births)) # 변수 DataSet 설정
          # list를 zip 함수로 변수 names와 births를 묶는다 
df = pd.DataFrame(data = DataSet, columns = ['Names', 'Births']) # DataFrame 생성
```

### * 행 단위 데이터 추출하기

##### - 데이터를 행 단위로 추출하려면 loc, iloc 속성을 사용한다.
##### - loc : 인덱스를 기준으로 행 데이터 추출, 파이썬에서 리스트 자료형에 저장된 데이터 순서를 인덱스라 한다.
##### - iloc : 행 번호를 기준으로 행 데이터 추출, 판다스에서는 행 번호로 부른다.

```python
# loc 속성에 대괄호를 이용하여 인덱스를 전달하면 인덱스에 해당하는 행의 데이터를 추출한다
print(df.loc[0])
```

```python
# print(df.loc[-1]) # loc 속성에서 [-1]은 인덱스 범위에 포함되지 않는다.
```

```python
# iloc 속성에 1을 전달하여 데이터를 추출한다.
print(df.iloc[1])
```

```python
# iloc 속성은 loc 속성과 달리 음수를 사용해도 데이터 추출이 가능하다.
# -1은 마지막 행 데이터를 추출한다.
print(df.iloc[-1])
```

```python
# 인덱스 0, 3, 5인 데이터를 한번에 추출한다.
print(df.loc[[0,3,5]])
```

### * tali 메서드

##### - 데이터프레임의 마지막 행 데이터를 추출하는 또 다른 방법으로 tail()를 사용한다.
##### - tail()의 인자 n에 1을 전달하면 마지막 행의 데이터를 추출한다.

```python
print(df.tail(n=1)) # 제일 마지막 값을 출력
```

#### tail()와 loc속성이 반환하는 데이터의 자료형은 다르다
##### - loc 속성이 반환한 데이터 자료형은 시리즈이다.
##### - tail()가 반환한 데이터 자료형은 데이터프레임이다.

```python
sub_loc = df.loc[0]  # sub_loc 변수 선언
sub_tail = df.tail(n=1) # sub_tail 변수 선언

print(type(sub_loc)) # 데이터타입 : Series
print(type(sub_tail)) # 데이터타입 : DataFrame
```

```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table', header=0, index_col=0)
# read_html() : 웹 페이지에서 테이블 형태의 데이터를 추출하는 함수

summer = df[1].iloc[1:, :5]
summer.columns = ['경기수', '금', '은', '동', '계']
summer.sort_values('금', asciicending=False) # ascending : 
```

##### ① 마스크 기능은 특정한 조건을 만족하는지에 따라 참과 거짓을 반환한다.
##### ② 행과 열을 바꾼 데이터프레임

```python
import pandas as pd
import numpy as np
index = pd.date_range('2020/1/1', periods=8) # date_range('시작일', 기간=8)
df = pd.DataFrame(np.random.rand(8,3), index= index, columns = list('ABC'))

print(df['B'])
```

```python
# ① 마스크 기능은 특정한 조건을 만족하는지에 따라 참과 거짓을 반환한다.
import pandas as pd
import numpy as np
index = pd.date_range('2020/1/1', periods=8) # date_range('시작일', 기간=8)
df = pd.DataFrame(np.random.rand(8,3), index= index, columns = list('ABC'))

print(df['B'] > 0.4) # B행 중에서 0.4보다 큰지에 따라 참과 거짓을 출력한다.
```

```python
df2 = df[df['B'] > 0.4]
df2
```

```python
# ② 행과 열을 바꾼 데이터프레임
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(8,3), index= index, columns = list('ABC'))
df.T # .T를 사용하면 행과 열을 바꾸어서 추출할 수 있다.
```

```python
df2 = df2.T
df2
```
