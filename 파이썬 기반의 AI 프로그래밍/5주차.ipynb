### * 다층 퍼셉트론(MLP: Multipayer perceptrons)
- 여러 단계를 거쳐 결정을 만들어내는 선형 모델의 일반화 형태
- 입력 특성과 예측은 노드, 계수는 노드 사이의 연결로 나타낸 로지스틱 회귀
- 가중치 합을 만드는 과정 여러번 반복
- 중간 단계를 구성하는 은닉 유닛 계산 후 최종 결과 산출 위해 다시 가중치 합 계산
- 많은 계수(가중치) 학습
- 여러개의 가중치 합 계산 = 하나의 가중치 합 계산

### * 신경망 튜닝
- 신경망 복잡도 제어 : 학습 시작 전 가중치 무작위 설정 -> 모델 학습에 영향
- 복잡도 제어 : 은닉층 수, 은닉층의 유닛 갯수, 규제(alpha)
- 대량 데이터 기반 정보 획득과 복잡한 모델 생성 가능
- 오랜 학습 시간
- 데이터 전처리

```python
# 신경망 튜닝 실습

import graphviz
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import matplotlib.pyplot as plt
import platform
```

```python
from sklearn.datasets import make_moons
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import mglearn

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=y, random_state=42)

mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10])
mlp.fit(X_train, y_train)

mglearn.plots.plot_2d_separator(mlp, X, fill=True, alpha=.4)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.show()
```

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10,10]) # 은닉층
mlp.fit(X_train, y_train)

mglearn.plots.plot_2d_separator(mlp, X, fill=True, alpha=.3)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.show()
```

```python
# MLP에서는 L2규제를 적용한다.
# 비선형 함수. 기본적으로 생략하면 렐루 함수 적용
mlp = MLPClassifier(solver='lbfgs', activation='tanh', random_state=0, hidden_layer_sizes=[10,10])
mlp.fit(X_train, y_train)

mglearn.plots.plot_2d_separator(mlp, X, fill=True, alpha=.3)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.show()
```

```python
# 신경망 : L2규제
# 가중치를 0에 가깝게 감소시켜서 모델의 복잡도 제시
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

fig, axes = plt.subplots(2, 4, figsize=(20, 8))
for axx, n_hidden_nodes in zip(axes, [10, 100]):
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[n_hidden_nodes,n_hidden_nodes], alpha=alpha)
        mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, ax=ax, alpha=.3)
        mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
        ax.set_title("n_hidden=[{},{}]\n alpha={:.4f}".format(n_hidden_nodes, n_hidden_nodes, alpha))

plt.show()

# 메모리 에러 (RAM 부족)
# batch size 조절. 페이징 파일 크기 조절(SystemPropertiesAdvanced)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

print("유방암 데이터의 특성별 최대값 : {}".format(cancer.data.max(axis=0)))

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)

print("훈련 세트 정확도 : {:.2f}".format(mlp.score(X_train, y_train)))
print("테스트 세트 정확도 : {:.2f}".format(mlp.score(X_test, y_test)))
```

```python
mean_on_train = X_train.mean(axis=0)
std_on_train = X_train.std(axis=0)

X_train_scaled = (X_train-mean_on_train)/std_on_train
X_test_scaled = (X_test-mean_on_train)/std_on_train

mlp = MLPClassifier(random_state=0)
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도 : {:.2f}".format(mlp.score(X_train_scaled, y_train)))
print("테스트 세트 정확도 : {:.2f}".format(mlp.score(X_test_scaled, y_test)))
```

```python
mlp = MLPClassifier(random_state=0, max_iter=1000)
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled, y_train)))
print("테스트 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled, y_test)))
```

```python
mlp = MLPClassifier(random_state=0, alpha=1, max_iter=1000)
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도 : {:.3f}".format(mlp.score(X_train_scaled, y_train)))
print("테스트 세트 정확도 : {:.3f}".format(mlp.score(X_test_scaled, y_test)))
```

```python
plt.figure(figsize=(20, 5))
plt.imshow(mlp.coefs_[0], interpolation='none', cmap='viridis')
plt.yticks(range(30), cancer.feature_names)
plt.xlabel("은닉유닛")
plt.ylabel("입력특성")
plt.colorbar()
```

### * 분류 예측의 불확실성 추정
- 어떤 테스트 포인트에 대해 분류기가 예측한 클래스가 무엇인지 정확한 클래스임을 얼마나 확신하는 지가 중요한 경우에 발생한다.
- decision_function
- predict_proba


#### * predict_proba()에서 출력은 범위가 제한적이다.

```python
# 분류 예측의 불확실성 추정 실습
import matplotlib.pyplot as plt
import platform
```

```python
# 분류 예측에 대한 결과에 대해서 확실한 것인지 아닌지에 평가하는 형태
# decision_function 사용
# predict_proba 사용
from sklearn.datasets import make_moons
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import mglearn
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(noise=0.25, factor=0.5, random_state=1)
y_named = np.array(["blue", "red"])[y]
X_train, X_test, y_train_named, y_test_named, y_train, y_test = train_test_split(X, y_named, y, random_state=0)

gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train_named)
```

```python
print("X_test.shape : ", X_test.shape)
print("결정 함수 결과 형태 : ", gbrt.decision_function(X_test).shape)
# (n_samples)
```

```python
print("결정 함수 : ", gbrt.decision_function(X_test)[:6])
# 양수일 때, 클래스 1에 포함된다.
# 음수일 때, 클래스 0에 포함된다.
```

```python
print("임계치와 결정 함수 결과 비교 : \n", gbrt.decision_function(X_test) > 0)
print("예측 : \n", gbrt.predict(X_test))
```

```python
greater_zero = (gbrt.decision_function(X_test) > 0).astype(int)
pred = gbrt.classes_[greater_zero]

print("pred 예측 결과와 같다 : ", np.all(pred == gbrt.predict(X_test)))
```

```python
decision_function = gbrt.decision_function(X_test)
print("결정 함수 최소값 : {:.2f}, maximum : {:.2f}".format(np.min(decision_function), np.max(decision_function)))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
mglearn.tools.plot_2d_separator(gbrt, X, ax=axes[0], alpha=.4, fill=True, cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(gbrt, X, ax=axes[1], alpha=.4, cm=mglearn.ReBl)

for ax in axes:
    mglearn.discrete_scatter(X_test[:, 0], X_test[:, 1], y_test, markers='^', ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, markers='o', ax=ax)
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 0")
    
cbar = plt.colorbar(scores_image, ax=axes.tolist())
cbar.set_alpha(1)
cbar.draw_all()
axes[0].legend(["테스트 클래스 0", "테스트 클래스 1", "훈련 클래스 0", "훈련 클래스 1"], ncol=4, loc=(.1, 1.1))
```

```python
print("확률 값의 형태 : ", gbrt.predict_proba(X_test).shape)
# (n_samples.2)
```

```python
print("예측 확률 : ")
# predict_proba 

print(gbrt.predict_proba(X_test[:6]))
# 복잡도가 낮은 모델은 예측의 불확실성이 높은 편이다.
# -> 불확실성과 모델의 정확도가 동등하도록 만들어서 이를 보정한다고 표현
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

mglearn.tools.plot_2d_separator(gbrt, X, ax=axes[0], alpha=.4, fill=True, cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(gbrt, X, ax=axes[1], alpha=.5, cm=mglearn.ReBl, function='predict_proba')

for ax in axes:
    mglearn.discrete_scatter(X_test[:, 0], X_test[:, 1], y_test, markers='^', ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, markers='o', ax=ax)
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
    
cbar = plt.colorbar(scores_image, ax=axes.tolist())
cbar.set_alpha(1)
cbar.draw_all()
axes[0].legend(["테스트 클래스 0", "테스트 클래스 1", "훈련 클래스 0", "훈련 클래스 1"], ncol=4, loc=(.1, 1.1))
```
