### - csv 파일에서 데이터 읽어오기
 - csv 모듈을 불러옴
 - open()함수로 열어서 f에 저장
 - f를 reader() 함수에 넣어 data라는 reader 객체 생성
 - data를 출력
 - f.close()로 파일 닫기

 ### - csv 파일 for 반복문을 사용하여 한 줄씩 읽어오기

```python
import csv
f = open('report.csv', 'r', encoding='UTF-8')
data = csv.reader(f)
for row in data:
    print(row)
```

### - 헤더 저장
 - 헤더(header)는 데이터 파일에서 여러가지 값들이 어떤 의미를 갖는지 표시한 행
 - 수많은 데이터를 다룰 때 표지판과 같은 역할
 - 헤더를 통해 각 열의 데이터가 어떤 의미를 갖는지 알 수 있기 때문에 매우 중요하다.

```python
import csv
f = open('weather.csv')
data = csv.reader(f)
header = next(data) # header 변수에 헤더 데이터 행을 저장
#print(header)       # header 변수를 출력

for row in data:
    print(row)

f.close()
```
